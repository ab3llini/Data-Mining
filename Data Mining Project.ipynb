We check if there is some trend in the sales considering all days of the week. In the first part we focus on all the store and we will make a further analysis on regional and store type trends


```python
import dataset.dataset as d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = d.read_dataset()
df['Date'] = df['Date'].astype("datetime64[s]")
df['day_of_week'] = df['Date'].dt.weekday_name
df_sales_per_day = df.groupby(df["day_of_week"])['NumberOfSales'].agg('sum')

df_sales_per_day.plot(kind="bar", title='Sales divided by Day of the Week' )
```

```python
df = d.read_dataset()
df['Date'] = df['Date'].astype("datetime64[s]")
df['day_of_week'] = df['Date'].dt.weekday_name
df_divided_by_month = df.groupby(df["Date"].dt.month)['NumberOfSales'].agg('sum')
df_divided_by_month.plot(kind="bar", title='Sales divided by Month' )
```
